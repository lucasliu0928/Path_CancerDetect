```python
import os
import shutil
```

```python
proj_dir = "/fh/fast/etzioni_r/Lucas/mh_proj/mutation_pred/intermediate_data/"
src_path = proj_dir + "/1_tile_pulling/OPX/IMSIZE250_OL100/"
dest_path = proj_dir + "4_tile_feature/OPX/IMSIZE250_OL100/"
folders = os.listdir(src_path)
folders.sort()
```

```python
# Loop through the range of OPX folders
for folder_name in folders:
    
    # Define source and destination paths
    src = f"{src_path}{folder_name}/features"
    dest = f"{dest_path}{folder_name}/features"
    
    # Check if the source directory exists
    if os.path.exists(src):
        # Create the destination directory if it doesn't exist
        os.makedirs(os.path.dirname(dest), exist_ok=True)
        
        # Move the features folder to the destination
        shutil.move(src, dest)
        
        print(f"Moved {src} to {dest}")
    else:
        print(f"Source directory {src} does not exist")
```
