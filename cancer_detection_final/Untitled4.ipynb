```python
import torch
import torch.nn as nn

embed_dim = 256
num_heads = 8
multihead_attn = nn.MultiheadAttention(embed_dim, num_heads, batch_first = True)
```

```python
query = torch.rand(1, 10, embed_dim)  # (batch_size, sequence_length, embed_dim)
key = torch.rand(1, 10, embed_dim)
value = torch.rand(1, 10, embed_dim)

attn_output, attn_output_weights = multihead_attn(query, key, value)
```
