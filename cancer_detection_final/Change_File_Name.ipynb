```python
import os
```

```python
input_path = '/fh/fast/etzioni_r/Lucas/mh_proj/mutation_pred/intermediate_data/cancer_prediction_results110224/IMSIZE250_OL100/'
pt_folders  = os.listdir(input_path)
pt_folders.sort()
pt_folders.remove('.DS_Store')
```

```python
old_file_name = 'features_alltiles_nonoverlapretccl.h5'
new_file_name = 'features_alltiles_retccl.h5'
```

```python
for folder_name in pt_folders:
    folder_path = os.path.join(input_path, folder_name, "features")
    old_file_path = os.path.join(folder_path, old_file_name)
    new_file_path = os.path.join(folder_path, new_file_name)
    # Check if the file exists in the folder
    if os.path.exists(old_file_path):
        os.rename(old_file_path, new_file_path)
        print(f"Renamed {old_file_path} to {new_file_path}")
    else:
        print(f"{old_file_path} not found in {folder_path}")
```
